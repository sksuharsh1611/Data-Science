```python
# Ans 3:
'''
The ArithmeticError class is a built-in Python exception class that represents errors that occur during arithmetic operations, 
such as division by zero or integer overflow. 

It is a subclass of the Exception class and a superclass of more specific arithmetic exception classes, 
such as ZeroDivisionError, OverflowError, FloatingPointError, and others.

'''
# for example : ZeroDivisionError
a = 10
b = 0
try:
    c = a / b
except ZeroDivisionError:
    print("Error: Division by zero")
```

```python
# For example 'AssertionError'
a = 10
b = 20
try:
    assert a > b, "a is not greater than b"
except AssertionError as e:
    print(str(e))
```

```python
# Ans 4:
'''
The LookupError class is used as a base class for exceptions that occur when a key or index is not found in a sequence or mapping object. 
This includes exceptions like IndexError and KeyError.

The KeyError exception is raised when you try to access a key that does not exist in a dictionary.
'''

# for example: raises a KeyError exception
d = {'a': 1, 'b': 2, 'c': 3}
try:
    value = d['d']  # KeyError: 'd' not in dictionary
except KeyError:
    print("Error: Key not found")
```

```python
# for example: raises an IndexError exception
l = [1, 2, 3]
try:
    value = l[3]  # IndexError: list index out of range
except IndexError:
    print("Error: Index out of range")
```

```python
# for example : In this example, the except block catches both KeyError and IndexError 
d = {'a': 1, 'b': 2, 'c': 3}
l = [1, 2, 3]
try:
    value = d['d']
    value = l[3]
except LookupError:
    print("Error: Key or index not found")
```

```python
# Ans 5:
'''
1. ImportError: It is an error that occurs when a Python module or package cannot be imported. 
                This can happen due to various reasons, such as 
                    i)   a missing or incorrect module name, 
                    ii)  a problem with the module's code, or 
                    iii) a problem with the Python environment.

2. ModuleNotFoundError: It is a specific type of ImportError that occurs when a module is not found. 
                        It is introduced in Python 3.6 to replace the previous behavior of raising an ImportError with the message "No module named module_name". 
                        The ModuleNotFoundError exception makes the error message more explicit and easier to understand.

'''
# for example:
try:
    import pytho
except ModuleNotFoundError:
    print("Error: Module not found")
```
