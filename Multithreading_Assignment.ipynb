```python
# Ans 1:
'''
a. Multithreading in Python is a way to run multiple threads of execution concurrently within a single process. 
   A thread is a lightweight subprocess that can run concurrently with other threads.

b. Multithreading is used to improve the performance of programs that can benefit from parallelism, 
   such as I/O-bound or CPU-bound tasks. 
   It allows for better resource utilization and can help reduce overall execution time.

c. The 'threading' module is used to handle threads in Python. 
'''
```

```python
# Ans 2:
'''
The threading module in Python is used to create and manage threads of execution within a program. 
It provides a high-level interface for working with threads and synchronization primitives such as locks and semaphores.

Uses of the following functions in the threading module:
a. activeCount(): This function returns the number of active threads in the current thread's thread object. 
                  It can be used to get a count of all the threads currently running in a program.

b. currentThread(): This function returns a reference to the current thread object. 
                    It can be used to get information about the current thread, such as its name, ID, and status.

c. enumerate(): This function returns a list of all active thread objects in the current thread's thread object. 
                It can be used to get information about all the threads running in a program, such as their name, ID, and status. 
                This function is particularly useful when debugging a program that uses multiple threads.
'''
```

```python
# Ans 3:
'''
a. run(): This method is called when a thread is started using the start() method. 
          It contains the code that will be executed in the thread.

b. start(): This method starts a new thread of execution for the run() method. 
            When the method is called, a new thread is created and the run() method is executed in that thread.

c. join(): This method blocks the calling thread until the thread on which it is called completes execution. 
            It can be used to ensure that a thread has completed its work before the program continues executing.

d. isAlive(): This method returns True if the thread is currently executing, and False otherwise. 
              It can be used to check the status of a thread and determine if it is still running.
'''
```

```python
# Ans 4: 
import logging
import threading
import time

# Configure logging
logging.basicConfig(filename='thread_logs.log', level=logging.DEBUG,
                    format='%(asctime)s - %(levelname)s - %(message)s')

# Function to calculate squares
def calc_squares(numbers):
    logging.debug("Calculating squares")
    for num in numbers:
        logging.debug(f"Square of {num}: {num*num}")

# Function to calculate cubes
def calc_cubes(numbers):
    logging.debug("Calculating cubes")
    for num in numbers:
        logging.debug(f"Cube of {num}: {num*num*num}")

if __name__ == "__main__":
    # Create list of numbers
    numbers = [1, 2, 3, 4, 5]

    # Create threads
    t1 = threading.Thread(target=calc_squares, args=(numbers,))
    t2 = threading.Thread(target=calc_cubes, args=(numbers,))

    # Start threads
    t1.start()
    t2.start()

    # Wait for threads to finish
    t1.join()
    t2.join()

    logging.debug("Threads have completed")
```

```python
# Ans 5:
'''
# Advantages of multithreading:

1. Increased performance by utilizing multiple CPU cores.
2. Improved responsiveness and user experience.
3. Ability to handle multiple tasks simultaneously.
4. Efficient use of system resources.
5. Simplifies complex problems by breaking them down into smaller, manageable tasks.

# Disadvantages of multithreading:

1. Increased complexity in design and debugging.
2. Difficulty in managing shared resources, leading to race conditions and deadlocks.
3. Potential for thread interference, leading to unpredictable behavior.
4. Increased overhead due to context switching between threads.
5. Difficult to predict and reproduce issues due to non-deterministic nature of thread execution.
'''
```

```python
# Ans 6:
'''
Deadlocks: It occurs when two or more threads are blocked waiting for each other to release resources they need to continue executing, resulting in a state where none of the threads can proceed. 
           Deadlocks can result in a system freeze or crash.

Race conditions: It occur when two or more threads attempt to access and modify shared resources simultaneously, resulting in unexpected or incorrect behavior. 
                 Race conditions can cause a variety of issues, such as 
                 a. data corruption ;
                 b. inconsistent program behavior ; 
                 c. can be difficult to debug and reproduce. 
            
        
Proper synchronization mechanisms, such as locks or semaphores, can help prevent race conditions.

'''
```
