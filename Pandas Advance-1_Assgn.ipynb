```python
# Ans 2:
# Define the index for the DataFrame
index = ['First','Second','Third']

# Create the DataFrame with columns 'A', 'B', and 'C', and the specified index
df = pd.DataFrame({"A":[1,2,3], "B":[5,6,7], "C":[8,9,10]}, index=index)

def reindex_dataframe(df):
    # Create a new index that starts from 1 and increments by 2 for each row
    new_index = pd.Index(range(1, 2*len(df), 2))
    
    # Set the new index on the DataFrame
    df = df.set_index(new_index)
    
    return df

new_df = reindex_dataframe(df)
print(new_df)
```

```python
# Sol 7:
# Define a Pandas DataFrame with an 'Email' column
df = pd.DataFrame({"Email":["suharshkumar@gmail.com",
                           "sudhanshu@yahoo.com",
                           "pwskills@iNeuron.com",
                           "john.doe@example.com"]
                  })

# Define a function that extracts the username from an email address and adds it as a new column
def extract_username(df):
    # Use the apply method to split each email address at the '@' symbol and extract the first part (the username)
    df['Username'] = df['Email'].apply(lambda x: x.split('@')[0])
    
    # Return the updated DataFrame with the new 'Username' column
    return df

# Call the extract_username function with the DataFrame as an argument and assign the result to a new variable
final_df = extract_username(df)

# Print the resulting DataFrame with the new 'Username' column
print(final_df)
```

```python
# Sol 8:
# Creating a dataframe with columns 'A', 'B', and 'C'
df = pd.DataFrame({"A":[3,8,6,2,9],
                  "B":[5,2,9,3,1],
                  "C":[1,7,4,5,2]
                  })

# Defining a function to select all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10.
def custom_dataframe(df):
    # Filtering the dataframe based on the conditions given using boolean indexing
    new_df = df[(df['A']>5) &(df['B'] <10)]
    return new_df

# Calling the function to get the filtered dataframe
final_df = custom_dataframe(df)

# Printing the final dataframe
print(final_df.loc[1:4:2,['A','B','C']])



```

```python
# Sol 9:

# Creating a pandas DataFrame with a single column 'Values'
df = pd.DataFrame({"Values":[4,5,6,8,2,1]})

# Defining a function to calculate the mean of a DataFrame column
def mean(df):
    return sum(df['Values']) / len(df['Values'])

# Defining a function to calculate the median of a DataFrame column
def median(df):
    sorted_values = sorted(df['Values'])  # sorting the values in ascending order
    length = len(df['Values'])  # getting the length of the column
    if length % 2 == 0:
        # if the length is even, return the average of the middle two values
        return (sorted_values[length//2] + sorted_values[length//2 - 1]) / 2
    else:
        # if the length is odd, return the middle value
        return sorted_values[length//2]

# Defining a function to calculate the standard deviation of a DataFrame column
def std_dev(df):
    n = len(df['Values'])  # getting the length of the column
    if n <= 1:
        # if the length is less than or equal to 1, the standard deviation is 0
        return 0
    mean_val = mean(df)  # calculating the mean of the column
    variance = sum((val - mean_val) ** 2 for val in df['Values']) / (n - 1)  # calculating the variance of the column
    return variance ** 0.5  # returning the square root of the variance to get the standard deviation

# Calculating the mean, median and standard deviation of the 'Values' column in the DataFrame
mean_val = mean(df)
median_val = median(df)
std_dev_val = std_dev(df)

# Printing the results
print(f'Mean: {mean_val}, Median: {median_val}, Standard Deviation: {std_dev_val}')
```

```python
# Sol 10:
import pandas as pd

# Create a DataFrame with columns 'Date' and 'Sales'
df = pd.DataFrame({
    'Date': pd.date_range(start='2022-01-01', end='2022-01-12'),
    'Sales': [10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65]
})

# Define a function that adds a new column 'MovingAverage' to a given DataFrame 'df'
def moving_average(df):
    # Calculate the moving average of the 'Sales' column using a window size of 7
    # and including the current day in the calculation
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
    # Return the DataFrame with the new 'MovingAverage' column added
    return df

# Call the 'moving_average' function and pass in the DataFrame 'df'
df = moving_average(df)

# Print the DataFrame 'df' with the new 'MovingAverage' column
print(df)
```

```python
# Sol 11:

# Create a DataFrame with dates from '2023-01-01' to '2023-01-05'
df = pd.DataFrame({"Date": pd.date_range(start='2023-01-01', end='2023-01-05')})

# Define a function to add a new column 'Weekday' to a given DataFrame 'df'
def show_weekdays(df):
    # Extract the weekday name from the 'Date' column and assign it to a new column 'Weekday'
    df['Weekday'] = df['Date'].dt.day_name()
    # Return the DataFrame with the new 'Weekday' column added
    return df

# Call the 'show_weekdays' function and pass in the DataFrame 'df'
df = show_weekdays(df)

# Print the DataFrame 'df' with the new 'Weekday' column
print(df)
```

```python
# Sol 12:

# Define a function to select custom dates from a DataFrame
def select_custom_dates(df):
    
    condition = (df['Date'] >= '2023-01-01') & (df['Date'] <= '2023-01-31')
    
    # Use loc function to select rows where the condition is True and return the resulting DataFrame
    return df.loc[condition]

# Create a DataFrame with dates from '2023-01-01' to '2023-12-31'
df = pd.DataFrame({"Date":pd.date_range(start='2023-01-01', end='2023-12-31')})

# Call the select_custom_dates function to select dates between '2023-01-01' and '2023-01-31'
df = select_custom_dates(df)

# Print the resulting DataFrame
print(df)
```
