```python
# Ans 2:
from flask import Flask

app = Flask(__name__)  #creating object of Flask server

@app.route("/hello_world")    #App-routing : Binding through function name
def hello_world():           #Creating method to display "Hello world" on web
    return "<h1>Hello World!!</h1>"       #using <h1> Tag printing "Hello World"

if __name__=="__main__":   #Creating main function 
    app.run(host="0.0.0.0")
    
# Output URL >> https://salmon-nurse-bgiox.pwskills.app:5000/hello_world
```

```python
# Ans 3:
'''
In Flask, app routing refers to the process of defining URLs or routes that map to specific functions or views in a Flask application. 
The app routing mechanism in Flask is used to determine how incoming HTTP requests should be handled and what response should be returned to the client.

In Flask, app routes are used to define the URLs that users can visit to access different parts of the application. 
For example, an app route could be defined to handle a request to the root URL of the application ("/"),
          or to handle requests to specific pages or resources such as "/login", "/logout", "/users", or "/api".

# We use app routes in Flask for the following reasons:

1. Routing: App routes allow us to define the URLs or routes that users can visit to access different parts of the application. This makes it easy to map incoming requests to specific functions or views that handle the request and generate a response.

2. URL Building: App routes also make it easy to generate URLs or links to different parts of the application. This is useful when creating links to other pages or resources within the application.

3. HTTP Methods: App routes also allow us to define which HTTP methods (GET, POST, PUT, DELETE, etc.) are allowed for a particular route. This helps to ensure that the correct method is used when interacting with different parts of the application.

4. Modularization: App routes also allow us to modularize our Flask application by separating different parts of the application into separate files or modules. This helps to keep the application organized and makes it easier to maintain and update over time.
'''
```

```python
# Ans 4:
from flask import Flask

app = Flask(__name__)  #creating object of Flask server

@app.route("/welcome")
def welcome():
    return "<h1><b>Welcome to ABC Corporation</b></h1>"

@app.route("/")
def details():
    CompanyName = 'ABC Corporation'
    Location = 'India'
    Contact_Detail = '999-999-9999'
    return '''Company Name:{}<br>
              Location:{}<br>
              Contact Detail:{}'''.format(CompanyName,Location,Contact_Detail)

if __name__=="__main__":
    app.run(host="0.0.0.0")

    
# URL for "/welcome" route >> https://salmon-nurse-bgiox.pwskills.app:5000/welcome
# URL for "/" route >> https://salmon-nurse-bgiox.pwskills.app:5000/
```

```python
# Ans 5:
'''
In Flask, the url_for() function is used for URL building. 
This function is used to generate a URL for a specific function or view in the Flask application.

The url_for() function takes the name of the function or view as its first argument 
and any additional keyword arguments for generating the URL. 

The keyword arguments correspond to the variable parts of the URL pattern and are used to replace the placeholders in the URL pattern with actual values.
'''

from flask import Flask, url_for

app = Flask(__name__)

# Define a function to handle requests to the root URL ("/")
@app.route('/')
def home():
    return 'Hello, World!'

# Define a function to handle requests to a dynamic route with a variable part
@app.route('/user/<name>')
def user(name):
    return f'Hello, {name}!'

# Define a function to handle requests to a named route
@app.route('/login')
def login():
    return 'Please login'

with app.test_request_context():
    # Generate a URL for the home() function
    home_url = url_for('home')
    print(home_url)

    # Generate a URL for the user() function with a value for the 'name' variable
    user_url = url_for('user', name='John')
    print(user_url)

    # Generate a URL for the login() function
    login_url = url_for('login')
    print(login_url)
```
